```python
from PIL import Image, ImageDraw, ImageFont
import matplotlib.pyplot as plt

# Open the image
imgE = Image.open("/content/meme1.png")

# Check image size and mode
print("Image Size:", imgE.size)
print("Image Mode:", imgE.mode)

# Initialize ImageDraw
DRAW = ImageDraw.Draw(imgE)

# Preparing the Text Elements
TEXT_1 = "**DATA SCIENCE"
TEXT_2 = "BEFORE INNOMATICS"
TEXT_3 = "AFTER INNOMATICS"

# Choose a font and size
# You may need to install or specify the path to a TTF font file if this doesn't work
font = ImageFont.truetype("/content/Inconsolata-VariableFont_wdth,wght.ttf", 20)

# Image dimensions
img_width, img_height = imgE.size

# Set the coordinates outside the main picture area
text_x_offset = img_width + 10  # Adding an offset to position text on the right side

# Coordinates for text placement on the right side
DRAW.text((text_x_offset, img_height * 0.1), TEXT_1, fill=(0, 0, 0), font=font, align="right")
DRAW.text((text_x_offset, img_height * 0.4), TEXT_2, fill=(255, 0, 0), font=font, align="left")
DRAW.text((text_x_offset, img_height * 0.7), TEXT_3, fill=(0, 255, 0), font=font, align="left")

# Expand the canvas to fit the new text
expanded_img = Image.new("RGB", (img_width + 200, img_height), "white")
expanded_img.paste(imgE, (0, 0))

# Draw the text on the expanded image
DRAW = ImageDraw.Draw(expanded_img)
# Text 1: Top-right of the image
DRAW.text((text_x_offset, 0), TEXT_1, fill=(0, 0, 0), font=font, align="right")
DRAW.text((text_x_offset, img_height * 0.1), TEXT_2, fill=(255, 0, 0), font=font, align="left")
DRAW.text((text_x_offset, img_height * 0.7), TEXT_3, fill=(0, 255, 0), font=font, align="left")

# Show the image with updated text
plt.imshow(expanded_img)
plt.axis('off')  # Hide axes for better visualization
plt.show()
```
